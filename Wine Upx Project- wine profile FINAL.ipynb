```python
# Convert the dictionary into DataFrame 
wine_df = pd.DataFrame(wine_ds2) 
   
# Before renaming the columns 
print(wine_df.columns) 
   
```

```python
wine_df.rename(columns = {'residual sugar':'Sweetness', 'chlorides':'Salty', 
                              'density':'Density'}, inplace = True) 
```

```python
from sklearn import preprocessing
```

```python
#OBSEVATION : Variables that contribute to Sourness are pH,fixed acidity', 'volatile acidity' and 'citric acid'
# Normalize Variabled that contribute to Sourness,INDIVIDUALLY 
#Adding the normanlied vvarable
```

```python
wine_df.head()
```

```python
#data["concavity_mean"]=((data["concavity_mean"]-data["concavity_mean"].min())/(data["concavity_mean"].max()-data["concavity_mean"].min()))*20

wine_df["pH"]=((wine_df["pH"]-wine_df["pH"].min())/(wine_df["pH"].max()-wine_df["pH"].min()))*1
```

```python
wine_df.head()
```

```python
wine_df["fixed acidity"]=((wine_df["fixed acidity"]-wine_df["fixed acidity"].min())/(wine_df["fixed acidity"].max()-wine_df["fixed acidity"].min()))*1
```

```python
wine_df.head()
```

```python
wine_df["volatile acidity"]=((wine_df["volatile acidity"]-wine_df["volatile acidity"].min())/(wine_df["volatile acidity"].max()-wine_df["volatile acidity"].min()))*1
```

```python
wine_df.head()
```

```python
wine_df["citric acid"]=((wine_df["citric acid"]-wine_df["citric acid"].min())/(wine_df["citric acid"].max()-wine_df["citric acid"].min()))*1
```

```python
wine_df.head()
```

```python
# adding all values of acidity and pH to get a single indicator for sourness
#drop column named 'Sourness" is a  summation of nomalised values of all variables that conibute to 'sour taste' 
```

```python
Sourness=wine_df['fixed acidity']+ wine_df['volatile acidity']+wine_df['citric acid']+wine_df['pH']
wine_df['Sourness']=Sourness
```

```python
wine_df.head()
```

```python
wine_df.drop(['fixed acidity', 'volatile acidity', 'citric acid','pH'], axis=1, inplace=True)
```

```python
wine_df.head()
```

```python
wine_df["free sulfur dioxide"]=((wine_df["free sulfur dioxide"]-wine_df["free sulfur dioxide"].min())/(wine_df["free sulfur dioxide"].max()-wine_df["free sulfur dioxide"].min()))*1
wine_df["total sulfur dioxide"]=((wine_df["total sulfur dioxide"]-wine_df["total sulfur dioxide"].min())/(wine_df["total sulfur dioxide"].max()-wine_df["total sulfur dioxide"].min()))*1
wine_df["sulphates"]=((wine_df["sulphates"]-wine_df["sulphates"].min())/(wine_df["sulphates"].max()-wine_df["sulphates"].min()))*1
```

```python
wine_df.head()
```

```python
Odour=wine_df['free sulfur dioxide']+ wine_df['total sulfur dioxide']+wine_df['sulphates']
wine_df['Odour']=Odour
```

```python
wine_df.head()
```

```python
wine_df.drop(['total sulfur dioxide','free sulfur dioxide','sulphates'],axis=1, inplace=True)
wine_df.head()
```

```python
wine_df.rename(columns = {'residual sugar':'Sweetness', 'chlorides':'Salty', 
                              'density':'Density', 'alcohol':'Alcohol','quality':'Quality'}, inplace = True)
wine_df.head()
```

```python
wine_df.describe().T
```

```python
wine_df["Sweetness"]=((wine_df["Sweetness"]-wine_df["Sweetness"].min())/(wine_df["Sweetness"].max()-wine_df["Sweetness"].min()))*1
wine_df["Alcohol"]=((wine_df["Alcohol"]-wine_df["Alcohol"].min())/(wine_df["Alcohol"].max()-wine_df["Alcohol"].min()))*1
wine_df.describe().T
```
